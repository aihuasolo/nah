Translate Test

```python
from transformers import pipeline
translate = pipeline(task="translation",model="t5-small")
text="translate English to German:I love China"
translate(text)
```

Sentiment-analaysis

```python
pipe = pipeline("sentiment-analysis")
pipe("今天北京真冷啊")
```

```python
pipe("Today beijing is so cold")
```

```python
text_list=['Today beijing is really cold.','I  think the taste if this is average','You learn things really quick']
pipe(text_list)
```

Token classification

```python
classifier = pipeline(task="ner")
preds  =classifier("Higging Face is a French company based in New York City")
print(preds)
```

QA

```python
q_a = pipeline(task="question-answering")
preds = q_a(
    question="what is the name of the repo?",
    context="The name of the repo is huggingface/transformers"
)
print(f"score:{round(preds['score'],4)},start:{preds['start']},end:{preds['end']},answer:{preds['answer']}")
```
